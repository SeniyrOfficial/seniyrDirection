```python
def subjectTally(subjects,outputLabels):
    for i in subjects:
        print(i)
        if( i == "Physics"):
            outputLabels[0] += 6
            outputLabels[1] += 4
        if( i == "Chemistry"):
            outputLabels[0] += 3
            outputLabels[1] += 5
        if( i == "Mathematics"):
            outputLabels[0] += 7
            outputLabels[4] += 4
            outputLabels[3] += 2
        if( i == "Business Studies"):
            outputLabels[4] += 5
            outputLabels[3] += 5
        if( i == "Literature"):
            outputLabels[2] += 6
            outputLabels[3] += 2
        if( i == "Accountancy"):
            outputLabels[4] += 6
            outputLabels[3] += 5
        if( i == "Biology"):
            outputLabels[1] += 10
            outputLabels[0] += 1
        if( i == "Computer Science"):
            outputLabels[0] += 7
        if( i == "Economics"):
            outputLabels[2] += 4
            outputLabels[3] += 6
            outputLabels[4] += 6
        if( i == "Psychology"):
            outputLabels[2] += 6
        if( i == "History"):
            outputLabels[2] += 6
        if( i == "Geography"):
            outputLabels[2] += 6
        if( i == "Political Science (Civics)"):
            outputLabels[2] += 6

    return outputLabels
```
