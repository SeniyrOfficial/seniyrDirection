```python

#created a bernoulli class
 
class seniyrDataProdUtils():
    def bernoulliPmf(x,p):
        """
        probability mass function        
        """
        f = p**x*(1-p)**(1-x)
        return f
    
    def bernoulliMean(p):
        """
        expected value of bernoulli random variable
        """
        return p
    
    def bernoulliVar(p):
        """
        variance of bernoulli random variable
        """
        return p*(1-p)
    
    def bernoulliStd(p):
        """
        standart deviation of bernoulli random variable
        """
        return seniyrDataprodUtils.bernoulliVar(p)**(1/2)
    
    def bernoulliRvs(p,size=1):
        """
        random variates
        """
        rvs = np.array([],dtype=int)
        for i in range(0,size):
            if np.random.rand() <= p:
                a=1
                rvs = np.append(rvs,a)
            else:
                a=0
                rvs = np.append(rvs,a)
        return rvs

    def ublbRandom(lb,ub):
        return lb + (ub - lb)*np.random.rand()

    def ublbRandint(lb,ub,size = 1):
        return np.random.randint(lb,ub,size)

    
```
